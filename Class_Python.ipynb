'클래스(Class)'라는 것은 '설계도'와 '설명서'가 합쳐진 형태이다.
- __init__() 등과 같은 것들을 통해 설명서를 작성하는 거임. 

```python
class BlackBox:
  pass
```

```python
b1 = BlackBox() # b1 객체는 BlackBox의 인스턴스이다.
b1.name = '까망이' # 변수 선언
print(b1.name)
```

```python
# 특정 객체가 어느 클래스의 인스턴스인지를 확인하는 함수
print(isinstance(b1, BlackBox))
```

```python
class BlackBox:
  def __init__(self, name, price):
    self.name = name
    self.price = price

# __init__(): 우리가 만든 BlackBox라는 클래스는 name과 price라는 두 개의 변수를 가지고 있는데, 객체를 만들 때 각각 name과 price를 넣어주는 초기화 역할을 함.
# __init__()은 객체가 생성될 때 자동으로 실행됨. 
# 이때 name과 price는 '멤버 변수'라고 부른다. 
```

```python
b1 = BlackBox('까망이', 200000)
b2 = BlackBox('하양이', 100000)

print(b1.name, b1.price)
print(b2.name, b2.price)
```

self는 객체 자기 자신을 의미함. 

```python
class BlackBox:
  def __init__(self, name, price):
    self.name = name
    self.price = price
  
  def set_travel_mode(self, min):
    print(f'{self.name}, {min}분 동안 여행 모드 ON') # 메서드 내에서는 self.name과 같은 형태로 멤버 변수를 사용해야 한다. 
```

```python
b1 = BlackBox('까망이', 200000)
b2 = BlackBox('하양이', 100000)

b1.set_travel_mode(20)
b2.set_travel_mode(10)
```

```python
# BlackBox를 맨 앞에 적으면 정의된 메서드의 self 인자에 객체명을 넣어줌. 

BlackBox.set_travel_mode(b1, 20)
BlackBox.set_travel_mode(b2, 10)

# 즉, self 인자에 들어갈 것은 객체명인데, 대부분의 경우 객제명에 메서드를 호출하는 방식으로 사용하는 상황에서
# self인자에 들어갔어야 할 객체명을 이미 알고 있기 때문에 self인자를 생략하고 나머지 인자를 전달값으로 넣어주는 것임
```

# 상속

```python
class BlackBox:
  def __init__(self, name, price):
    self.name = name
    self.price = price
```

```python
class TravelBlackBox(BlackBox):
    def set_travel_mode(self, min): # 여행 모드 시간(분)
        print(str(min) + '분 동안 여행 모드 ON')
```

```python
b1 = BlackBox('까망이', 200000)
b2 = TravelBlackBox('하양이', 100000)

b2.set_travel_mode(10)
# b1 객체는 set_travel_mode() 메서드를 사용할 수 없음
```

### 부모 클래스의 메서드를 상속받되, 멤버 변수가 추가되는 경우

```python
class TravelBlackBox(BlackBox):
    def __init__(self, name, price, sd):
        BlackBox.__init__(self, name, price) # 부모 클래스의 이름을 그대로 쓰는 경우에는 self 인자를 추가해야 함.
        self.sd = sd
        # name과 price는 BlackBox 클래스를 호출해서 설정하고, sd카드 부분만 새로이 인자를 초기화해서 객체를 생성.

    def set_travel_mode(self, min):
        print(str(min + '분 동안 여행 모드 ON'))
```

```python
# super를 사용

class TravelBlackBox(BlackBox):
    def __init__(self, name, price, sd):
        super().__init__(name, price) # super()를 이용하는 경우 self 인자를 사용하지 않음에 주의
        self.sd = sd
        # name과 price는 BlackBox 클래스를 호출해서 설정하고, sd카드 부분만 새로이 인자를 초기화해서 객체를 생성.

    def set_travel_mode(self, min):
        print(str(min + '분 동안 여행 모드 ON'))
```

```python
class Calculator:
    def __init(self):
        self.result = 0

    def add(self, num):
        self.result += num
        return self.result
    
cal1 = Calculator()
cal2 = Calculator()
```

```python
print(cal1.add(3))
print(cal1.add(4))
print(cal2.add(3))
print(cal2.add(7))
```
